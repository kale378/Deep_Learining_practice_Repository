```python
from mlxtend.plotting import plot_decision_regions   # plot_decision_regions is used for classifing the data.
```

```python
import tensorflow as tf
import numpy as np
import pandas as pd
from pylab import rcParams
import matplotlib.pyplot as plt
import warnings
from matplotlib.colors import ListedColormap
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles
import seaborn as sns
```

```python
# Make_circles is an dataset which is present in sklearn
# Here we draw it
X, y = make_circles(n_samples=100, noise=0.1, random_state=1)    # take the 100 samples, Noise = ??
                                                                # Here x and y shows that there are two variables in dataset
```

```python
sns.scatterplot(X[:,0],X[:,1],hue=y)     # the code creates a scatter plot using the values from the first and second columns 
                                          # of x, and assigns colors to the points based on the categories or groups represented
                                          # by the variable y.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2)
```

```python
model = Sequential()         #Sequential() represent here we will form a neural network

model.add(Dense(256, input_dim=2, activation='relu'))   # input neuron =2, one hidden layer with 256 neuron.
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history=model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=3500, verbose='auto')
#validation_data =The model will use this 
# validation data to evaluate its performance during the training process.
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

```python
plot_decision_regions(X_test, y_test.ravel(), clf = model, legend=2)
plt.show()
```
